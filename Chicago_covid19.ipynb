```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
pd.set_option('display.max_columns',500)

df = pd.read_csv("~/Downloads/chicago_covid.csv")
df.head()
```

```python
df.duplicated().sum()
```

```python
print(df.shape)
print(df.info())
```

```python
df.rename(columns = {'Cases - Weekly':'Weekly Cases',
                     'Cases - Cumulative':'Cumulative Cases',
                    'Case Rate - Weekly':'Weekly Case Rate (per 100,000)',
                    'Case Rate - Cumulative':'Cumulative Case Rate (per 100,000)',
                    'Tests - Weekly':'Weekly Tests',
                    'Tests - Cumulative':'Cumulative Tests',
                     'Test Rate - Weekly':'Weekly Test Rate (per 100,000)',
                     'Test Rate - Cumulative':'Cumulative Test Rate (per 100,000)',
                    'Percent Tested Positive - Weekly':'Weekly Positivity Percent',
                    'Percent Tested Positive - Cumulative':'Cumulative Positivity Percent',
                    'Deaths - Weekly':'Weekly Deaths',
                    'Deaths - Cumulative':'Cumulative Deaths',
                    'Death Rate - Weekly':'Weekly Death Rate (per 100,000)',
                    'Death Rate - Cumulative':'Cumulative Death Rate (per 100,000)'},inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df.dropna(subset = ['Weekly Tests','ZIP Code Location'],inplace = True)
```

```python
df = df.fillna(value = 4)
```
